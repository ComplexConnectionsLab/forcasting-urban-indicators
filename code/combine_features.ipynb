```python
import numpy as np
import pandas as pd
from pandas import IndexSlice as idx
import matplotlib.pyplot as plt
```

```python
wage = pd.read_csv("../data/processed/unstable/wage.csv", index_col=[0,1])
estab = pd.read_csv("../data/processed/unstable/estab.csv", index_col=[0,1])
emp = pd.read_csv("../data/processed/unstable/emp.csv", index_col=[0,1])
```

```python
areas = sorted(list(set.intersection(set([index[0] for index in wage.index.values]), set([index[0] for index in emp.index.values]), set([index[0] for index in estab.index.values]))))
wage = wage.loc[idx[areas, ],]
estab = estab.loc[idx[areas, ],]
emp = emp.loc[idx[areas, ],]
wage
```

```python
features = {"wage":wage, "emp":emp, "estab":estab}
X = pd.concat(features, axis=1)
```

```python
X.to_csv("../data/processed/unstable/intersection_all.csv")
```

```python
wage = pd.read_csv("../data/processed/SAMI/wage.csv", index_col=[0,1])
estab = pd.read_csv("../data/processed/SAMI/estab.csv", index_col=[0,1])
emp = pd.read_csv("../data/processed/SAMI/emp.csv", index_col=[0,1])
areas = sorted(list(set.intersection(set([index[0] for index in wage.index.values]), set([index[0] for index in emp.index.values]), set([index[0] for index in estab.index.values]))))
```

```python
wage = pd.read_csv("../data/processed/unstable/wage.csv", index_col=[0,1])
estab = pd.read_csv("../data/processed/unstable/estab.csv", index_col=[0,1])
emp = pd.read_csv("../data/processed/unstable/emp.csv", index_col=[0,1])
```

```python
union_areas = sorted(list(set.union(*[set([index[0] for index in df.index.values]) for df in [wage, emp, estab]])))
union_areas = sorted(list(set.intersection(set(union_areas), set(areas))))
len(union_areas)
```

```python
wage = pd.read_csv("../data/processed/SAMI/wage.csv", index_col=[0,1]).loc[idx[union_areas,],]
estab = pd.read_csv("../data/processed/SAMI/estab.csv", index_col=[0,1]).loc[idx[union_areas,],]
emp = pd.read_csv("../data/processed/SAMI/emp.csv", index_col=[0,1]).loc[idx[union_areas,],]
```

```python
features = {"wage":wage, "emp":emp, "estab":estab}
X = pd.concat(features, axis=1)
X.to_csv("../data/processed/unstable/union_all.csv")
```
