```python
def stripStop(x):
    return x.strip(".")
```

```python
pops = {}
for file in listdir("../data/raw/pop/2000-2010/"):
    pops[file] = pd.read_excel(f"../data/raw/pop/2000-2010/{file}")
    pops[file] = pops[file].iloc[3:pops[file].shape[0]-8]
    pops[file].columns = ["Geographic Area", "April 1, 2000", "2000", "2001", "2002", "2003", "2004",
                          "2005", "2006", "2007", "2008", "2009", "April 1, 2010", "July 1, 2010"]
    state = pops[file]["Geographic Area"].iloc[0]
    pops[file] = pops[file].iloc[1:]
    pops[file]["Geographic Area"] += f", {state}"
    pops[file]["Geographic Area"] = pops[file]["Geographic Area"].apply(stripStop)
    
```

```python
pops
```

```python
counties = pd.concat(pops.values())
counties
```

```python
counties.to_csv("../data/raw/pop/pop_2000_2010.csv")
```
