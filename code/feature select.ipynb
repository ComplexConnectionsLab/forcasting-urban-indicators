```python
import numpy as np
import pandas as pd
from pandas import IndexSlice as idx
import matplotlib.pyplot as plt
from os import listdir
```

```python
path = "../data/raw/"
years = {}
for year in range(2000, 2020):
    years[year] = pd.read_csv(path+f"{year}/allhlcn{str(year%100).zfill(2)}.csv")
```

```python
for year in range(2000,2020):
    years[year] = years[year][years[year]["Area Type"] == "County"]

years[2000]
```

```python
for year in range(2000,2020):
    years[year] = years[year][years[year]["Ownership"] == "Total Covered"]

years[2000]
```

```python
for year in range(2000,2020):
    years[year] = years[year][["Area\nCode", "Area", "Annual Average Establishment Count", "Annual Average Employment", "Annual Total Wages",
                                 "Employment Location Quotient Relative to U.S.", "Total Wage Location Quotient Relative to U.S."]]

years[2000]
```

```python
for year in range(2000,2020):
    years[year].dropna(inplace=True)

years[2000]
```

```python
for year in range(2000,2020):
    years[year].drop(["Employment Location Quotient Relative to U.S.", "Total Wage Location Quotient Relative to U.S."], axis=1, inplace=True)

years[2000]
```

```python
areas = set.intersection(*[set(years[year]["Area"].values.tolist()) for year in range(2000, 2020)])
len(areas)
```

```python
for year in range(2000,2020):
	years[year].rename(columns={"Area\nCode":"Area Code"}, inplace=True)

years[2000]
```

```python
for year in range(2000,2020):
    years[year].set_index(["Area Code", "Area"], inplace=True)
    years[year] = years[year].loc[idx[:, list(areas)], :].sort_index()

years[2000]
```

```python
years[2000].index.values
```

```python
estab = pd.DataFrame(np.nan, index=pd.MultiIndex.from_tuples(years[2000].index.values, names=["Area Code", "Area"]), columns=list(range(2000, 2020)))
for index in years[2000].index.values:
    for year in range(2000,2020):
        estab.loc[index, year] = years[year].loc[index, "Annual Average Establishment Count"]

estab
```

```python
estab.isnull().values.any()
```

```python
estab = estab.astype(np.int64)
estab.to_csv("../data/processed/estab.csv")
```
