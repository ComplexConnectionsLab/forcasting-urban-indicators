```python
import pandas as pd
from pandas import IndexSlice as idx
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor
import warnings
warnings.filterwarnings("ignore")
```

```python
combined = pd.read_csv("../data/processed/unstable/intersection_all.csv", index_col=[0,1], header=[0,1])
combined
```

```python
X_train = combined.loc[:, idx[:, [str(i) for i in range(2000,2005)]]]
y_train = combined.loc[:, idx[:, "2005"]]
lin_reg = MultiOutputRegressor(LinearRegression()).fit(X_train, y_train)
print(f"Train Score: {lin_reg.score(X_train, y_train)}")
print(f"Coefs: {[model.coef_ for model in lin_reg.estimators_]}")
print(f"Intercepts: {[model.intercept_ for model in lin_reg.estimators_]}")
```

```python
for year in range(2001, 2015):
	X_test = combined.loc[:, idx[:, [str(i) for i in range(year,year+5)]]]
	y_test = combined.loc[:, idx[:, str(year+5)]]
	print(f"{year}-{year+5}: {lin_reg.score(X_test, y_test)}")
```

```python
X_train = combined.loc[:, idx[:, [str(i) for i in range(2000,2005)]]]
y_train = combined.loc[:, idx[:, [str(i) for i in range(2005,2007)]]]
lin_reg = MultiOutputRegressor(LinearRegression()).fit(X_train, y_train)
print(f"Train Score: {lin_reg.score(X_train, y_train)}")
print(f"Coefs: {[model.coef_ for model in lin_reg.estimators_]}")
print(f"Intercepts: {[model.intercept_ for model in lin_reg.estimators_]}")
```

```python
for year in range(2001, 2014):
	X_test = combined.loc[:, idx[:, [str(i) for i in range(year,year+5)]]]
	y_test = combined.loc[:, idx[:, [str(i) for i in range(year+5,year+7)]]]
	print(f"{year}-{year+6}: {lin_reg.score(X_test, y_test)}")
```

```python
X_train = combined.loc[:, idx[:, [str(i) for i in range(2000,2005)]]]
y_train = combined.loc[:, idx[:, [str(i) for i in range(2005,2010)]]]
lin_reg = MultiOutputRegressor(LinearRegression()).fit(X_train, y_train)
print(f"Train Score: {lin_reg.score(X_train, y_train)}")
print(f"Coefs: {[model.coef_ for model in lin_reg.estimators_]}")
print(f"Intercepts: {[model.intercept_ for model in lin_reg.estimators_]}")
```

```python
for year in range(2001, 2011):
	X_test = combined.loc[:, idx[:, [str(i) for i in range(year,year+5)]]]
	y_test = combined.loc[:, idx[:, [str(i) for i in range(year+5,year+10)]]]
	print(f"{year}-{year+9}: {lin_reg.score(X_test, y_test)}")
```

```python
diff = combined.diff(axis=1, periods=2).shift(periods=-1, axis=1) / 2
diff = diff.drop(columns=["2000", "2019"], level=1)
```

```python
diffdiff = diff.diff(axis=1, periods=2).shift(periods=-1, axis=1) / 2
diffdiff = diffdiff.drop(columns=["2001", "2018"], level=1)
```

```python
X_diff_train = combined.loc[:, idx[:, [str(i) for i in range(2000,2005)]]]
y_diff_train = diff.loc[:, idx[:, "2005"]]
diff_lin_reg = MultiOutputRegressor(LinearRegression()).fit(X_diff_train, y_diff_train)
print(f"Train Score: {diff_lin_reg.score(X_diff_train, y_diff_train)}")
```

```python
X_diffdiff_train = combined.loc[:, idx[:, [str(i) for i in range(2000,2005)]]]
y_diffdiff_train = diffdiff.loc[:, idx[:, "2005"]]
diffdiff_lin_reg = MultiOutputRegressor(LinearRegression()).fit(X_diffdiff_train, y_diffdiff_train)
print(f"Train Score: {diffdiff_lin_reg.score(X_diffdiff_train, y_diffdiff_train)}")
```

```python
diff_grad_reg = MultiOutputRegressor(GradientBoostingRegressor()).fit(X_diff_train, y_diff_train)
print(f"Train Score: {diff_grad_reg.score(X_diff_train, y_diff_train)}")
```

```python
for year in range(2001, 2014):
	X_diff_test = combined.loc[:, idx[:, [str(i) for i in range(year,year+5)]]]
	y_diff_test = diff.loc[:, idx[:, str(year+5)]]
	print(f"{year}-{year+5}: {diff_grad_reg.score(X_diff_test, y_diff_test)}")
```

```python
diffdiff_grad_reg = MultiOutputRegressor(GradientBoostingRegressor()).fit(X_diffdiff_train, y_diffdiff_train)
print(f"Train Score: {diffdiff_grad_reg.score(X_diffdiff_train, y_diffdiff_train)}")
```

```python
for year in range(2003, 2013):
	X_diffdiff_test = diffdiff.loc[:, idx[:, [str(i) for i in range(year,year+5)]]]
	y_diffdiff_test = diffdiff.loc[:, idx[:, str(year+5)]]
	print(f"{year}-{year+5}: {diffdiff_grad_reg.score(X_diffdiff_test, y_diffdiff_test)}")
```
