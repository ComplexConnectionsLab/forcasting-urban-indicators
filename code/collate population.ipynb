```python
import numpy as np
import pandas as pd
```

```python
pop0 = pd.read_csv("../data/raw/pop/pop_2000_2010.csv")
pop0 = pop0.drop(["Unnamed: 0"], axis=1).set_index("Geographic Area")
pop0.head()
```

```python
pop0 = pop0.drop(["April 1, 2000", "April 1, 2010", "July 1, 2010"], axis=1)
pop0.head()
```

```python
pop1 = pd.read_csv("../data/raw/pop/pop_2010_2019.csv")
pop1 = pop1.drop(["Unnamed: 0"], axis=1).set_index("Geographic Area")
pop1.head()
```

```python
pop0.sort_index(inplace=True)
pop1.sort_index(inplace=True)
```

```python
cities = set.intersection(set(pop0.index.values.tolist()), set(pop1.index.values.tolist()))
pop0 = pop0.drop(set(pop0.index.values.tolist()).difference(cities))
pop1 = pop1.drop(set(pop1.index.values.tolist()).difference(cities))
print(pop0.shape[0])
print(pop1.shape[0])
```

```python
pop = pd.concat([pop0, pop1], axis=1)
pop
```

```python
pop.to_csv("../data/processed/pop.csv")
```
