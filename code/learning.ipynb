```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['text.usetex'] = True
mpl.rcParams.update(mpl.rcParamsDefault)
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

## 1 ESTAB

```python
df = pd.read_csv("../data/processed/SAMI/estab.csv", index_col=[1,0])
df
```

```python
sns.lineplot(data=(df.iloc[::int(df.shape[0]/10)]).T)
plt.legend(loc='center right', bbox_to_anchor=(1.6, 0.5), ncol=1)
plt.xticks(rotation=45, ha='right')
plt.title("Sample of Establishments Time Series")
plt.show()
```

```python
X = df[[str(i) for i in range(2000, 2010)]]
y = df[[str(i) for i in range(2018, 2020)]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1729)
reg = MultiOutputRegressor(GradientBoostingRegressor()).fit(X_train, y_train)
reg.score(X_test, y_test)
```

## 2 EMP

```python
df = pd.read_csv("../data/processed/SAMI/emp.csv", index_col=[1,0])
df
```

```python
sns.lineplot(data=(df.iloc[:100:10]).T)
plt.legend(loc='center right', bbox_to_anchor=(1.6, 0.5), ncol=1)
plt.xticks(rotation=45, ha='right')
plt.title("Sample of Employment Time Series")
plt.show()
```

```python
X = df[[str(i) for i in range(2000, 2010)]]
y = df[[str(i) for i in range(2010, 2020)]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1729)
reg = MultiOutputRegressor(GradientBoostingRegressor()).fit(X_train, y_train)
reg.score(X_test, y_test)
```

```python
reg.predict(X.iloc[90].to_numpy().reshape(1,-1)) - y.iloc[90].to_numpy().reshape(1,-1)
```

```python
X = df[[str(i) for i in range(2000, 2019)]]
y = df[[str(i) for i in range(2019, 2020)]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1729)
lin_reg = LinearRegression().fit(X_train, y_train)
print(f"Train Score: {lin_reg.score(X_train, y_train)}")
print(f"Test Score: {lin_reg.score(X_test, y_test)}")
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
X = df[[str(i) for i in range(2000, 2018)]]
y = df[[str(i) for i in range(2018, 2020)]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1729)
lin_reg = MultiOutputRegressor(LinearRegression()).fit(X_train, y_train)
print(f"Train Score: {lin_reg.score(X_train, y_train)}")
print(f"Test Score: {lin_reg.score(X_test, y_test)}")
```
